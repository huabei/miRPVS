```python
from utils import read_dock_score
```

```python
origin_100K_data_path = '../data/3a6p/100k'
origin_10K_data_path = '../data/3a6p/10k_apH'
# origin_10K_data_path = '../data/3a6p/10k'

dock_score_100k = dict(read_dock_score(origin_100K_data_path))
dock_score_10k = dict(read_dock_score(origin_10K_data_path))
```

```python
dock_1 = list()
dock_2 = list()
for key, value in dock_score_10k.items():
    if value == -16.816:
        print(key)
        continue
    dock_1.append(dock_score_100k[key])
    dock_2.append(value)
```

```python
from package import plot_fit_confidence_bond
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(dock_1, dock_2)
```

```python
# import matplotlib
# matplotlib.use('Qt5Agg')
fig = plot_fit_confidence_bond(x=dock_1, y=dock_2, r2=r2, annot=False)
fig.savefig('repeat_dock_10k.png')
```
